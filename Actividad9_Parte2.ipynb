## üìå Paso 1.Generar un mapa **base** para luego agregar los puntos

```python
import folium
```

```python
#Creamos un mapa base centrado en la provincia
mapa = folium.Map(location=[-50.0, -70.0], zoom_start=6)
mapa
```

## üìå Paso 2. Definir la funci√≥n para  agregar los puntos

```python
#funcion que agrega los puntos en el mapa
def agregar_puntos(df, mapa):
    for i, fila in df.iterrows():
        folium.Marker(
            location=[fila["latitud"], fila["longitud"]],
            popup=f"ID: {fila['gid']} - Distrito: {fila['nombre_del_distrito']}",
            icon=folium.Icon(color="blue", icon="train", prefix="fa")
        ).add_to(mapa)
```

## üìå Paso 3. Agrergar los puntos al mapa

```python
#Filtramos santa cruz
santa_cruz = ferro[ferro["nombre_del_distrito"] == "SANTA CRUZ"]
```

```python
#Agregamos puntos en el mapa y mostramos
agregar_puntos(santa_cruz, mapa)
mapa
```
